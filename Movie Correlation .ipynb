```python
#Import libraries :
import pandas as pd
import numpy as np
#import matplotlip
import matplotlib.pyplot as plt
```

```python
#Read Data :
df = pd.read_csv('C:/Users/c/Downloads/movies.csv', sep = ',')
```

```python
df.head()
```

```python
# Let's See if there is Missing Value
for col in df.columns :
    missing_value = np.mean(df[col].isnull())
    print("{} - {}%".format(col,round(missing_value,3)))
```

```python
# Data Type : 
df.dtypes
# Change Data Type Of Some Columns : 
df['votes'] = df['votes'].astype('Int64')
df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
df['runtime'] = df['runtime'].astype('Int64')
```

```python
df["budget"].dtypes
```

```python
#df.fillna(0, inplace= True)
for col_name in ["votes", "budget", "gross","runtime"]:
    if (True in list(df[col_name].isnull().unique())):
        mean = int(df[col_name].mean())
        print(col_name, " => ", mean)
        df[col_name].fillna(mean, inplace= True)
```

```python
df
```

```python
df = df.dropna().reset_index(drop=True)
```
